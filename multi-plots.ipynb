```python
from scipy import interpolate
def confidence_intervals(z,dx=1,dy=1):
    """
    takes normalised probability distribution as input (fi dx,dy=1 then probdist=n/Ntotal)
    outputs the level of the contour at which 75% 50% and 25%
    """
    n=20
    t = np.linspace(0, z.max(), n)
    integral = ((z >= t[:, None, None]) * z).sum(axis=(1,2)) *dx*dy
    f = interpolate.interp1d(integral, t)
    t_contours = f(np.array([0.95,0.68,0.40,0.20]))
    return np.append(t_contours,1)

```

```python
def easyplot(valx,valy,valp,edgex,edgey):
    lvl = confidence_intervals(valx.T/valx.sum())
    ax.contourf(edgex,edgey,valx.T/valx.sum(),vmin=0,vmax=vmax,levels=lvl,cmap='Purples',alpha=alpha)
    c1 = ax.contour(edgex,edgey,valx.T/valx.sum(),vmin=0,vmax=vmaxcont,levels=lvl,cmap='Purples',linewidths=3)
    lvl = confidence_intervals(valy.T/valy.sum())
    ax.contourf(edgex,edgey,valy.T/valy.sum(),vmin=0,vmax=vmax,levels=lvl,cmap='Oranges',alpha=alpha)
    c2 = ax.contour(edgex,edgey,valy.T/valy.sum(),vmin=0,vmax=vmaxcont,levels=lvl,cmap='Oranges',linewidths=3)
    lvl = confidence_intervals(valp.T/valp.sum())
    CS = ax.contour(edgex,edgey,valp.T/valp.sum(),vmin=0,vmax=vmaxcont,levels=lvl,cmap='binary',linewidths=4,linestyles='dashed')
    for c in CS.collections:
        c.set_dashes([(1, (3.0, 3.0))])
    h1,l1 = c1.legend_elements()
    h2,l1 = c2.legend_elements()
    h3,l3 = CS.legend_elements()
    return h1,h2,h3
```
